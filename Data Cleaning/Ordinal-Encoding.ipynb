```python
import pandas as pd
```

```python
df = pd.DataFrame({'Size' : ['L', 'S', 'xL', 'M', 'M', 'L', 'xL', 'S', 'M', 'xS']})
```

```python
df
```

## Encosing using skLearn

```python
ord_unique = [['xS', 'S', 'M', 'L', 'xL']]
```

```python
ord_unique
```

```python
from sklearn.preprocessing import OrdinalEncoder

oe = OrdinalEncoder(categories = ord_unique)

df['sk_encoded'] = oe.fit_transform(df[['Size']]).astype(int)
```

```python
df
```

## Encosing using map()

```python
ord_unique = {
    'xS' : 0,
    'S' : 1,
    'M' : 2,
    'L' : 3,
    'xL' : 4,
}

df['map_encoded'] = df['Size'].map((ord_unique))
```

```python
df
```
