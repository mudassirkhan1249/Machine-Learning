# Identifying and Handling Missing Values

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"my_file (1).csv")
df.head(3)
```

```python
df.shape
```

## FINDING MISSING VALUE

```python
df.isnull().sum()
```

```python
(df.isnull().sum() / df.shape[0]) * 100
```

```python
df.isnull().sum().sum()
```

```python
(df.isnull().sum().sum() / (df.shape[0] * df.shape[1])) * 100
```

```python
df.notnull().sum()
```

```python
df.notnull().sum().sum()
```

```python
plt.figure(figsize = (6,4))
sns.heatmap(df.isnull())
plt.show()
```

## HANDLING MISSING VALUE

```python
df.head(3)
```

```python
df.isnull().sum()
```

```python
df[['Peak', 'All Time Peak']].dtypes
```

```python
df['All Time Peak'].fillna(df['All Time Peak'].mode()[0])
```

```python
for i in df.select_dtypes(object).columns:
    df[i].fillna(df[i].mode()[0], inplace = True)

df.head(3)
```
