```python
from sklearn.impute import SimpleImputer
si = SimpleImputer(strategy = 'most_frequent')
```

```python
col = df.select_dtypes(object).columns
col
```

```python
arr = si.fit_transform(df[col])
obj_df = pd.DataFrame(arr, columns = col)
```
